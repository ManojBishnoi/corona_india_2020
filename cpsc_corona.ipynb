```python
# removing interval 

i =0
lit = []
for x in list_age1:
    if len(x) == 2:
        lit.append(x)
    else:
        # taking average of intervals
        lit.append((int(x[:2])+int(x[-2:]))/2) 
# conversion to integer         
myList = [int(i) for i in lit]
print(myList)
len(myList)
```

```python
#basic histogram to check corona virus with the age factor.
fig= plt.figure(1, figsize=(50, 20))
#adding one subplot
ax=fig.add_subplot(111)
plt.hist(myList,bins=50, facecolor='orange', edgecolor = 'blue')
plt.xlim(0,80)
plt.xlabel('Age',FontSize = 45)
plt.ylabel('Detected Cases',FontSize = 45)
plt.xticks(range(0,101,10))
```
